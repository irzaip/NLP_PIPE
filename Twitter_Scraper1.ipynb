# GET TWEET FROM SEARCH

```python
from twitterscraper import query_tweets

query = 'Keke'
worker = 10

list_of_tweets = query_tweets(query, worker)

#Or save the retrieved tweets to file:
def write_to_file(output_file):
    file = open(output_file,"w",encoding='UTF8')
    for tweet in query_tweets(query, worker):
      file.write(str(tweet.text.encode('utf-8')))
    file.close()
#write_to_file("output.txt")
```
